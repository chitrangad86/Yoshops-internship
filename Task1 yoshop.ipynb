```python
# import xlsxwriter module
import xlsxwriter

# Workbook() takes one, non-optional, argument
# which is the filename that we want to create.
workbook = xlsxwriter.Workbook('test.xlsx')

# The workbook object is then used to add new
# worksheet via the add_worksheet() method.
worksheet = workbook.add_worksheet()

# Use the worksheet object to write
# data via the write() method.
worksheet.write('A1', 'Emp. Code')
worksheet.write('B1', 'Emp name')
worksheet.write('C1', 'Father name')
worksheet.write('D1', 'Mother Name')
worksheet.write('D1', 'Address')

row = 1
column = 1
#input data 
Name = ["ankit", "rahul", "priya", "harshita",
                    "sumit", "neeraj", "shivam"]
 
# iterating through content list
for item in Name :
 
    # write operation perform
    worksheet.write(row, column, item)
 
    # incrementing the value of row by one
    # with each iterations.
    row += 1

Emp_Code = [1,2,3,4,5,6,7]
row = 1
column = 0
for item in Emp_Code:
 
    # write operation perform
    worksheet.write(row, column, item)
 
    # incrementing the value of row by one
    # with each iterations.
    row += 1

    
# again define the row and column values

row = 1
column = 2
#input data 
parent = (
    ['ankit', 'ankita'],
    ['rahul',   'ruchi'],
    ['priyank',  'priya'],
    ['harsh', 'harshita'],
    ['ajay', 'alka'],
    ["vijay","vijaya"],
    ["krishna","radha"]
)

for father, mother in parent:
    worksheet.write(row,column, father)
    worksheet.write(row, column+1, mother)
    row+=1

```

```python
import docx2txt
import re


my_text = docx2txt.process("CV_chitrangad_Jain.docx")

mail = re.compile(r'[a-zA-Z0-9-\.]+@[a-zA-Z-\.]*\.(com|edu|net)')
mo = re.compile(r'(?:(?:\+?([1-9]|[0-9][0-9]|[0-9][0-9][0-9])\s*(?:[.-]\s*)?)?(?:\(\s*([2-9]1[02-9]|[2-9][02-8]1|[2-9][02-8][02-9])\s*\)|([0-9][1-9]|[0-9]1[02-9]|[2-9][02-8]1|[2-9][02-8][02-9]))\s*(?:[.-]\s*)?)?([2-9]1[02-9]|[2-9][02-9]1|[2-9][02-9]{2})\s*(?:[.-]\s*)?([0-9]{4})(?:\s*(?:#|x\.?|ext\.?|extension)\s*(\d+))?')
mailid = mail.finditer(my_text)
mobile = mo.finditer(my_text)
for match in mailid:
    id=match.group(0)
    print(id) 
for match in mobile:
    no=print(match.group(0))
    print(no)
```
